```python
keep_cols = [
'loan_amnt','term','int_rate','installment',
'grade','sub_grade',
'emp_length','home_ownership','annual_inc','verification_status',
'purpose','addr_state',
'dti','delinq_2yrs',
'fico_range_low','fico_range_high',
'inq_last_6mths','open_acc','pub_rec',
'revol_bal','revol_util','total_acc',
'initial_list_status','issue_d',
'loan_status'
]

df = df[keep_cols]
```

```python
df = df.dropna()
df.head(15)
```

```python
df.shape
```

```python
cols_to_keep = [
'loan_amnt','term','int_rate','installment','grade','sub_grade',
'emp_length','home_ownership','annual_inc','verification_status',
'dti','fico_range_low','fico_range_high','inq_last_6mths',
'open_acc','pub_rec','revol_bal','revol_util','total_acc',
'loan_status'
]

df = df[cols_to_keep]
```

```python
df.info()
```

```python
df["loan_amnt"].head(15)
```

```python
df['loan_status'].value_counts()
```

```python
# remove running loans
df = df[df['loan_status'] != 'Current']
```

```python
# define bad categories
bad_status = [
    'Charged Off',
    'Default',
    'Late (31-120 days)',
    'Late (16-30 days)',
    'In Grace Period'
]

# create target
df['target'] = df['loan_status'].apply(lambda x: 1 if x in bad_status else 0)
```

```python
df['target'].head(10)
```

```python
df['target'].value_counts()
```

```python
df["target"].shape
```

```python
df.info()
```

```python
df['target'].value_counts(normalize=True)
```

```python
num_df = df.select_dtypes(include='number')
```

```python
num_df.info()
```

```python
corr_with_target = num_df.corr()['target'].sort_values(ascending=False)
corr_with_target
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(num_df.corr(), cmap='coolwarm', center=0)
plt.title("Correlation Matrix")
plt.show()
```

```python
Interest rate and FICO score showed the strongest relationship with default. 
Higher rates, higher DTI, and more credit inquiries increased risk, while strong credit scores reduced it.
In coreltion matrix the negative values are Variables that make borrower SAFER(negative correlation).
And , Positive corelations are are Variables that make borrower Unsafe.
```

```python
0 = GOOD borrower

1 = BAD / risky borrower
```

```python
grade_risk = df.groupby('grade')['target'].mean().sort_index()
grade_risk
```

```python
grade_risk.plot(kind='bar')
plt.ylabel("Default Rate")
plt.title("Default Rate by Grade")
plt.show()
```

```python
In lending, default simply means:

👉 the borrower failed to repay as agreed.
```

```python
df['income_bucket'] = pd.qcut(df['annual_inc'], 5)
```

```python
income_risk = df.groupby('income_bucket', observed=True)['target'].mean()
income_risk
```

```python
income_risk.plot(kind='bar')
plt.ylabel("Default Rate")
plt.title("Default Rate by Income Group")
plt.xticks(rotation=45)
plt.show()
```

```python
term_risk = df.groupby('term')['target'].mean()
term_risk
```

```python
term_risk.plot(kind='bar')
plt.ylabel("Default Rate")
plt.title("Default Rate by term")
plt.xticks(rotation=45)
plt.show()
```

```python
income_risk.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
#plt.ylabel("") # Hide y-label for cleaner look 
plt.title("Default Rate by Income Group") 
plt.show()
```

```python
df['dti_bucket'] = pd.qcut(df['dti'], 5)

dti_risk = df.groupby('dti_bucket',observed=True)['target'].mean()
dti_risk
```

```python
DTI ---- How much of a borrower’s income is already committed to paying debts.
DTI = Monthly Debt Payments / Monthly Income
```

```python
dti_risk.plot(kind='line', marker='o')
plt.ylabel("Default Rate")
plt.title("Default Rate by dti_bucket")
plt.xticks(rotation=45)
plt.show()
```

```python
Exploratory Data Analysis (EDA) – Summary

Exploratory Data Analysis was conducted to understand borrower characteristics, validate data quality, and identify relationships between applicant attributes and loan default behaviour.

The dataset (originating from LendingClub) was first cleaned by removing active loans and mapping repayment outcomes into a binary target variable where 0 represents fully paid loans and 1 represents default or delinquency.

Initial analysis showed an overall default rate of ~21%, reflecting a realistic class imbalance commonly observed in consumer lending portfolios.
```
